```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
leetcode_specific_question_infor= pd.read_csv('leetcode_specific_question_infor.csv')
leetcode_specific_question_infor.head()
```

```python
leetcode_specific_question_infor.columns
```

```python
leetcode_very_good_questions = leetcode_specific_question_infor[['questionFrontendId', 'title', 'difficulty', 'likes', 'dislikes', 'similarQuestions']][(leetcode_specific_question_infor['likes'] > leetcode_specific_question_infor['dislikes'] * 2) & (leetcode_specific_question_infor['likes'] > 1000)].sort_values(by=['likes'], ascending=False)
leetcode_very_good_questions
```

```python
leetcode_very_good_questions.to_csv(path_or_buf='leetcode_very_good_questions.csv')
```
